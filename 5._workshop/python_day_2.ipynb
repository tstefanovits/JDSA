# PREDICTIVE ANALYTICS

```python
import numpy.polynomial.polynomial as poly
```

```python
testdata=pd.Series([15,36,41,14,69,73,92,56,101,120,175,191,215,306,241,392])
```

```python
x=testdata.index
y=testdata.values
```

```python
testdata.plot()
```

```python
coefs = np.polyfit(x,y,2)
model = np.poly1d(coefs)
```

```python
x_test = np.linspace(0,16)
y_pred = model(x_test[:,None])
plt.plot(x,y)
plt.plot(x_test,y_pred,c='r')
plt.show()
```

```python
model(18)
```
